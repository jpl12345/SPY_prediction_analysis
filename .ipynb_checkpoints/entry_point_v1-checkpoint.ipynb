Rule-based Performance Measurement:
The rule-based performance is calculated by following a strategy of buying SPY when the rules match for a particular month. At the end of each month, the closing price of SPY is recorded. The sum of the closing prices for the months where the rules matched is computed, and the average is calculated by dividing the total by the number of months where the rules matched. This average price is then compared to the year-end price of the following year to evaluate the performance.

Reference Performance Measurement:
The reference performance is determined by benchmarking against a strategy of buying SPY every month throughout the year. At the end of each month, the closing price of SPY is recorded. The average price is computed by dividing the sum of the monthly closing prices by 12 (the number of months in a year). This average price is then compared to the year-end price of the following year to assess the performance.

These performance measurements provide a comparison between the rule-based strategy and the reference strategy, allowing for evaluation of the effectiveness of the rule-based approach.


```python
# Get the price of SPY at the end of each calendar year
yearly_prices_atEOY = {}

for year in spy_data['Date'].dt.year.unique():
    year_data = spy_data[spy_data['Date'].dt.year == year]
    year_end_price = year_data.iloc[-1]['Close']
    yearly_prices_atEOY[year] = year_end_price

print(yearly_prices_atEOY)
del yearly_prices_atEOY[2023] # 2023 is not over yet as of writing this code
print(yearly_prices_atEOY)

yearly_prices_atEOY_df = pd.DataFrame.from_dict(yearly_prices_atEOY, orient='index')
yearly_prices_atEOY_df = yearly_prices_atEOY_df.rename(columns={0: "EOY Price"})
yearly_prices_atEOY_df
```

part1:

```python
month_end_prices_whereRuleMatched = {}

for index, row in common_rows.iterrows():
    month_end_date = pd.Timestamp(year=row['Year'], month=row['Month'], day=1) + pd.offsets.MonthEnd(0)
    month_end_data = spy_data[spy_data['Date'] == month_end_date]
    if not month_end_data.empty:
        month_end_price = month_end_data['Close'].values[0]
        month_end_prices_whereRuleMatched[month_end_date] = month_end_price

print(month_end_prices_whereRuleMatched)
```

```python
# Calculate average price for each year and record the number of rule matches
yearly_average_prices = {}
rule_match_counts = {}

for date, price in month_end_prices_whereRuleMatched.items():
    year = date.year

    if year in yearly_average_prices:
        yearly_average_prices[year].append(price)
        rule_match_counts[year] += 1
    else:
        yearly_average_prices[year] = [price]
        rule_match_counts[year] = 1

# Calculate average price for each year and add to the DataFrame
for year, prices in yearly_average_prices.items():
    average_price = sum(prices) / len(prices)
    rule_match_counts[year] = {'Rule Matches': rule_match_counts[year], 'Yearly Average Price': average_price}

# Convert rule match counts to a DataFrame
rule_match_df = pd.DataFrame.from_dict(rule_match_counts, orient='index')

print("Rule Match Counts and Yearly Average Prices:")
print(rule_match_df)

try:
    rule_match_df = rule_match_df.drop(2023)
except KeyError:
    pass

```

```python
for year_where_rule_matched in rule_match_df_index:
  
    next_year = year_where_rule_matched + 1

    if (next_year == 2023):
        break
    
    print(f'{year_where_rule_matched} -> {rule_match_df.loc[year_where_rule_matched, "Yearly Average Price"]}')
    print(f'{next_year} -> {yearly_prices_atEOY_df.loc[next_year, "EOY Price"]}')
    yoy_change = yearly_prices_atEOY_df.loc[next_year, "EOY Price"] - rule_match_df.loc[year_where_rule_matched, "Yearly Average Price"]
    yoy_change_percent = ((yearly_prices_atEOY_df.loc[next_year, "EOY Price"] - rule_match_df.loc[year_where_rule_matched, "Yearly Average Price"])/yearly_prices_atEOY_df.loc[next_year, "EOY Price"])*100
    print(f'% change = {yoy_change_percent}')
    print()
```
